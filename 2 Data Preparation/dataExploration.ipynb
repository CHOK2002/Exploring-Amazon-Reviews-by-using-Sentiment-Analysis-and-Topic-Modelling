```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

def detect_stopwords(text_data, num_stopwords=10):
    stop_words = set(stopwords.words('english'))
    detected_stopwords = set()

    # count stopwords
    for text in text_data:
        words = word_tokenize(text.lower())
        for word in words:
            if word in stop_words:
                detected_stopwords.add(word)
                # stop if we have detected num_stopwords unique stopwords
                if len(detected_stopwords) == num_stopwords:
                    break
        if len(detected_stopwords) == num_stopwords:
            break

    stopword_string = ', '.join(detected_stopwords)
    return stopword_string

text_data = df['reviewText'].astype(str).tolist()
stopword_string = detect_stopwords(text_data)
print(f"Detected Stopwords: {stopword_string}")
```
